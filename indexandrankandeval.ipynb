```
import numpy as np
import pandas as pd

def index(files,bag):
    d=dict()
    for i in range(len(files)):
        for w in bag[i]:
            if w in d:
                d[w].add(files[i])
            else:
                d[w]=set()
                d[w].add(files[i])
    return d    

files=[]
bag=[]                   
files=df['docName']
bag=[x.split() for x in df['text']]

```

```
d=index(files,bag)
```

```
freq=[]
for i in d:
  freq.append([i,len(d[i])])
```

```
freq.sort(key=lambda x:x[1],reverse=True)
freq=freq[0:4000]
f=[freq[i][0] for i in range(len(freq))]
freq=f
```

```

words=list(d.keys())

words.sort()
```

```
words=words[5900:len(words)]
w=[]
for i in range(len(words)):
  if len(words[i])>=3 and words[i] in freq:
    w.append(words[i])
words=w
print(len(words))  
```

```
din=dict()
for i in d:
  if i in words:
    din[i]=d[i]
```

```
len(din)
```

```
from collections import Counter

words=list(din.keys())
words.sort()

#files*words matrix
mat=[[0 for i in range(len(din))] for j in range(len(files))]
#weights are tfidf values
for i in range(len(files)):
    tfd=Counter(bag[i])
    #for normalizing tf term
    mxtf=tfd.most_common(1)[0][1]
    for j in range(len(words)):
        if words[j] in bag[i]:
            #normalize tf 
            tf=tfd[words[j]]/mxtf
            #idf -> (tot docs )/(docss in which word is present)
            idf=np.log10(len(files))/len(din[words[j]])
            mat[i][j]=tf*idf
            
            
            
df=pd.DataFrame(mat)
```

```
import pandas as pd
import sklearn
import scipy
s=pd.read_csv('/content/drive/MyDrive/sem9/IRpackagenonsubmitfolder/matrix.csv',encoding='utf-8')
df=pd.read_csv('/content/drive/MyDrive/sem9/IRpackagenonsubmitfolder/clean.csv')

v=s.values
#Call scipy.spatial.distance.cosine(u, v) to find the cosine distance between vectors u and v.
# Subtract the result from 1 to find cosine similarity.
sel=df[df['label']==1]
relevant=sel['id']
cs=[]
for i in range(len(v)):
  cosine_similarity = 1-scipy.spatial.distance.cosine(v[0], v[i])
  cs.append([df['id'][i],cosine_similarity])

cs.sort(key=lambda x:x[1],reverse=True)
cs=[x[0] for x in cs]
ret=(cs[0:64])
relret = set(relevant).intersection(set(ret))
print('precision@R using cosine:',len(relret)/len(relevant))

```
